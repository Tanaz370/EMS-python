```python
# Install NumPy (Run this in terminal or notebook cell with !)
# !pip install numpy

import numpy as np

# 1. Create a 1D NumPy array with integers from 1 to 20
arr = np.arange(1, 21)
print("1D Array:", arr)

# a. Sum, Mean, Median, Std Dev
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))

# b. Indices of elements greater than 10
indices = np.where(arr > 10)
print("Indices where elements > 10:", indices)

# 2. Create a 2D array of shape 4x4 with numbers 1-16
arr2d = np.arange(1, 17).reshape(4, 4)
print("\n2D Array:\n", arr2d)

# a. Print array
# Already printed above

# b. Transpose
print("Transpose:\n", arr2d.T)

# c. Row-wise and Column-wise sum
print("Row-wise Sum:", np.sum(arr2d, axis=1))
print("Column-wise Sum:", np.sum(arr2d, axis=0))

# 3. Two 3x3 random arrays with integers 1-20
arr_a = np.random.randint(1, 21, (3, 3))
arr_b = np.random.randint(1, 21, (3, 3))
print("\nArray A:\n", arr_a)
print("Array B:\n", arr_b)

# a. Element-wise addition, subtraction, multiplication
print("Addition:\n", arr_a + arr_b)
print("Subtraction:\n", arr_a - arr_b)
print("Multiplication:\n", arr_a * arr_b)

# b. Dot product
print("Dot Product:\n", np.dot(arr_a, arr_b))

# 4. Reshape 1D array (size 12) into 3x4 and slice
arr_1d = np.arange(1, 13)
arr_reshaped = arr_1d.reshape(3, 4)
print("\nReshaped Array (3x4):\n", arr_reshaped)
print("First 2 rows and last 2 columns:\n", arr_reshaped[:2, -2:])
```

```python
# Install Pandas
# !pip install pandas

import pandas as pd

# 1. Create DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}
df = pd.DataFrame(data)

# a. Print first five rows
print(df.head())

# b. Summary statistics for Age & Salary
print(df[['Age', 'Salary']].describe())

# c. Average salary of HR employees
avg_hr_salary = df[df['Department'] == 'HR']['Salary'].mean()
print("Average Salary in HR:", avg_hr_salary)

# 2. Add Bonus column (10% of Salary)
df['Bonus'] = df['Salary'] * 0.10
print("\nDataFrame with Bonus:\n", df)

# 3. Filter employees aged between 25 and 30
filtered = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print("\nEmployees aged 25-30:\n", filtered)

# 4. Group by Department and average salary
dept_avg_salary = df.groupby('Department')['Salary'].mean()
print("\nAverage Salary by Department:\n", dept_avg_salary)

# 5. Sort by Salary and save to CSV
sorted_df = df.sort_values(by='Salary', ascending=True)
print("\nSorted DataFrame:\n", sorted_df)
sorted_df.to_csv('sorted_employees.csv', index=False)
```

```python
# Install Matplotlib
# !pip install matplotlib

import matplotlib.pyplot as plt
import numpy as np

# 1. Simple line plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.figure(figsize=(6,4))
plt.plot(x, y, marker='o')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()

# 2. Bar Graph
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks, color=['blue','green','orange','purple'])
plt.title('Marks of Students')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()

# 3. Pie Chart
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = [0.1 if r == max(revenue) else 0 for r in revenue]
plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Revenue Distribution')
plt.show()

# 4. Histogram
data = np.random.randint(1, 101, 1000)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title('Frequency Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
