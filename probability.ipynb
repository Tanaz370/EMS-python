```python
import random

# Simulate tossing a coin 10,000 times
trials = 10000
heads = 0
tails = 0

for _ in range(trials):
    toss = random.choice(['H', 'T'])
    if toss == 'H':
        heads += 1
    else:
        tails += 1

print("Experimental Probability of Heads:", heads / trials)
print("Experimental Probability of Tails:", tails / trials)
```

```python
# Rolling two dice 10,000 times
trials = 10000
count_sum_7 = 0

for _ in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    if die1 + die2 == 7:
        count_sum_7 += 1

print("Probability of getting sum = 7:", count_sum_7 / trials)
```

```python
def probability_at_least_one_six(trials=10000):
    success = 0
    for _ in range(trials):
        rolls = [random.randint(1, 6) for _ in range(10)]
        if 6 in rolls:
            success += 1
    return success / trials

print("Estimated Probability:", probability_at_least_one_six())
```

```python
colors = ['red'] * 5 + ['green'] * 7 + ['blue'] * 8
trials = 1000
prev_blue = 0
red_given_blue = 0

previous_ball = None
for _ in range(trials):
    ball = random.choice(colors)
    if previous_ball == 'blue':
        prev_blue += 1
        if ball == 'red':
            red_given_blue += 1
    previous_ball = ball

# Conditional probability
P_red_given_blue = red_given_blue / prev_blue if prev_blue else 0
print("P(Red | Previous Blue):", P_red_given_blue)
```

```python
import numpy as np

# Distribution: P(X=1)=0.25, P(X=2)=0.35, P(X=3)=0.4
values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]

sample = np.random.choice(values, size=1000, p=probabilities)

mean = np.mean(sample)
variance = np.var(sample)
std_dev = np.std(sample)

print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Exponential distribution with mean = 5 -> lambda = 1/mean
samples = np.random.exponential(scale=5, size=2000)

# Plot histogram and PDF
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')

# Overlay PDF
x = np.linspace(0, max(samples), 1000)
pdf = (1/5) * np.exp(-x/5)
plt.plot(x, pdf, 'r', linewidth=2)

plt.title('Exponential Distribution (mean=5)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Step a: Generate 10,000 random numbers from uniform distribution
population = np.random.uniform(0, 1, 10000)

# Step b: Draw 1000 samples of size n=30 and compute means
sample_means = []
for _ in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

# Plot original distribution
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
plt.hist(population, bins=30, color='skyblue', edgecolor='black')
plt.title('Uniform Distribution')

# Plot sample mean distribution
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='orange', edgecolor='black')
plt.title('Distribution of Sample Means (n=30)')

plt.show()
```
